```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Q1 -Ans

```python
# Given Data_Set
Data_Set = {
    'companies': ['Allied Signal', 'Bankers Trust', 'General Mills', 'ITT Industries', 'J.P.Morgan & Co.',
                        'Lehman Brothers', 'Marriott', 'MCI', 'Merrill Lynch', 'Microsoft', 'Morgan Stanley',
                        'Sun Microsystems', 'Travelers'],
    'Measure X': [24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26, 32.95, 91.36, 25.99, 39.42]
}
```

```python
# Create a DataFrame
df = pd.DataFrame(Data_Set)
```

```python
# Plot the data by using barplot (EDA)
plt.figure(figsize=(8, 5))
plt.bar(df['companies'], df['Measure X'])
plt.xticks(rotation=90)
plt.title('Measure X for Each Company')
plt.xlabel('companies')
plt.ylabel('Measure X (%)')
plt.show()
```

```python
# Calculate mean, standard deviation, and variance
mean_value = np.mean(df['Measure X'])
std_deviation = np.std(df['Measure X'], ddof=1)  # Use ddof=1 for sample standard deviation
variance = np.var(df['Measure X'], ddof=1)
```

```python
# Identify outliers using Tukey's method
Q1 = df['Measure X'].quantile(0.25)
Q3 = df['Measure X'].quantile(0.75)
IQR = Q3 - Q1                          #applying inter quartile range formula
 
lower_extreme = Q1 - 1.5 * IQR
upper_extreme = Q3 + 1.5 * IQR

outliers = df[(df['Measure X'] < lower_extreme) | (df['Measure X'] > upper_extreme)]
```

```python
# Print results
print(f"Mean (μ): {mean_value}")
print(f"Standard Deviation (σ): {std_deviation}")
print(f"Variance (σ^2): {variance}")
print("\nIdentified Outliers:")
print(outliers)
```
